```python
import time
import tensorflow as tf
i0 = tf.constant(0)
m0 = tf.zeros(shape=[0], dtype=tf.int32)
c = lambda i, m: i < 1000
b = lambda i, m: [i+1, tf.concat([m, i[tf.newaxis]], axis=0)]
i0, m0 = tf.while_loop(
    c, b, loop_vars=[i0, m0],
    shape_invariants=[i0.get_shape(), tf.TensorShape([None])], parallel_iterations=100)
with tf.Session() as sess:
    time_start = time.time()
    aa         = sess.run(m0)
    print(aa)
    time_end   = time.time()
    print(time_end-time_start)
```

```python
import tensorflow as tf
#tf.while_loop 各变量自身后一次迭代依赖于前一次迭代，各变量之间的迭代相互并行，但循环变量总是先行于其他变量的迭代
#变量之间的相互依赖关系会极大地降低并行度，因为第(i+1)个thread的变量(y+x)必须要等i个thread把其所依赖的变量(x)处理完，才可以进行后续的处理
#若各变量之间的相互依赖关系很小甚至没有(仅存在自身迭代的相互依赖关系)，那么第(i+1)个thread的变量(y)和第(i)个thread的变量(x)互不干扰，可以并行执行
#也就是说各个并行的thread都是先把自己能处理完的运算，即不需要依赖关系的运算，先处理完，剩下的需要依赖关系的，只好等前一个thread处理完，自己再处理剩下的运算
#也就是说无依赖关系的大量并行处理，有依来关系的只好顺序执行。
n = 10000
x = tf.constant(list(range(n)))
y = tf.constant(list(range(n)))
z = tf.constant(list(range(n)))
c = lambda i, x, y: i < n
#b = lambda i, x, y: (tf.Print(i + 1, [i]), tf.Print(x + 1, [i], "x:"), tf.Print(y + 1, [i], "y:"))
#b = lambda i, x, y: (tf.Print(i + 1, [i]), tf.Print(x + 1, [i], "x:"), tf.Print(y + x, [i], "y+x:"))
#b = lambda i, x, y: (tf.Print(i + 1, [i]), tf.Print(z + 1, [i], "x:"), tf.Print(z + 1, [i], "y:"))
def b(i, x, y):
    return tf.Print(i + 1, [i]), tf.Print(z + 1, [i], "x:"), tf.Print(z + 1, [i], "y:")
i, x, y = tf.while_loop(c, b, (0, x, y), parallel_iterations=100)
with tf.Session() as sess:
    #print(sess.run(i))  # prints [0] ... [9999]

    # The following line may increment the counter and x in parallel.
    # The counter thread may get ahead of the other thread, but not the
    # other way around. So you may see things like
    # [9996] x:[9987]
    # meaning that the counter thread is on iteration 9996,
    # while the other thread is on iteration 9987
    #print(sess.run(out).shape)
    sess.run([x, y])
```
